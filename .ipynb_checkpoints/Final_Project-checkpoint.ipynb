```python
#load data from xls file and transfer it into a dataframe
Data_1 = pd.ExcelFile('Paper1_figure_3G.xls')
df1 = Data_1.parse('Sheet1', skiprows=2, index_col=None)

#replace the empty cell with 0 in the data sheet
df1.fillna(0,inplace=True)
group11 = []
group12 = []
group13 = []

for ii in range(len(df1.values)):
    #skip data whose value equals 0
    if df1.values[ii][0] != 0:
        group11.append(df1.values[ii][0])
    if df1.values[ii][1] != 0:
        group12.append(df1.values[ii][1])
    if df1.values[ii][2] != 0:
        group13.append(df1.values[ii][2])

print(group11)
print(group12)

#Perform Mannâ€“Whitney Test
u_statistic_11, pVal_11 = stats.mannwhitneyu(group11, group12, alternative = 'two-sided')
print(u_statistic_11, pVal_11)
u_statistic_12, pVal_12 = stats.mannwhitneyu(group11, group13, alternative = 'two-sided')
print(u_statistic_12, pVal_12)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load data from xls file and transfer it into a dataframe
Data_3 = pd.ExcelFile('Paper1_figure_5C.xls')
```

```python
df = Data_3.parse('Sheet1', skiprows=2, na_values=['NA'])
```

```python
print(df)
```

```python
df.fillna(0,inplace=True)
print(df)
```

```python
group_labels = []
```

```python
for i in range(len(df.values)):
    group_labels.append(df.values[i][0])
```

```python
print(group_labels)
```

```python
group_21 = []
group_15 = []
group_21_v = []
group_15_v = []
group_significance = []
```

```python
for i in range(len(df.values)):
    group_temp1 = []
    group_temp2 = []
    
    for ii in range(1,5):
        if df.values[i][ii] != 0 :
            group_temp1.append(df.values[i][ii])
            
    for ii in range(5,9):
        if df.values[i][ii] != 0:
            group_temp2.append(df.values[i][ii])
    
    t_statistic, pVal = stats.ttest_ind(group_temp1, group_temp2)
    group_significance.append(pVal)
    
    group_21.append(np.mean(group_temp1))
    group_15.append(np.mean(group_temp2))
    
    group_21_v.append(stats.sem(group_temp1))
    group_15_v.append(stats.sem(group_temp2))
```

```python
print(group_significance)
```

```python
from matplotlib.markers import TICKDOWN

def significance_bar(start,end,height,displaystring,linewidth = 1.2,markersize = 8,boxpad  =0.3,fontsize = 15,color = 'k'):
    # draw a line with downticks at the ends
    plt.plot([start,end],[height]*2,'-',color = color,lw=linewidth,marker = TICKDOWN,markeredgewidth=linewidth,markersize = markersize)
    # draw the text with a bounding box covering up the line
    plt.text(0.5*(start+end),height,displaystring,ha = 'center',va='center',bbox=dict(facecolor='1.', edgecolor='none',boxstyle='Square,pad='+str(boxpad)),size = fontsize)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# the label locations
x = np.arange(len(group_labels))
# the width of the bars
width = 0.35 

# If we were to simply plot pts, we'd lose most of the interesting
# details due to the outliers. So let's 'break' or 'cut-out' the y-axis
# into two portions - use the top (ax1) for the outliers, and the bottom
# (ax2) for the details of the majority of our data
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True,gridspec_kw={'height_ratios':[1,5]})
fig.subplots_adjust(hspace=0.15)  # adjust space between axes

# plot the same data on both axes
rects1 = ax1.bar(x - width/2, group_21, width, label='21%$O_{2}$', color = 'k')
rects2 = ax1.bar(x + width/2, group_15, width, label='1.5%$O_{2}$',color = 'r')
rects1 = ax2.bar(x - width/2, group_21, width, color = 'k')
rects2 = ax2.bar(x + width/2, group_15, width, color = 'r')

# Add error bars for different groups of data
for pos, y, err in zip(x - width/2, group_21, group_21_v):
    ax1.errorbar(pos, y, err, lw = 1,
                capsize = 4, capthick = 2,
                color = "black")

for pos, y, err in zip(x + width/2, group_15, group_15_v):
    ax1.errorbar(pos, y, err, lw = 1,
                capsize = 4, capthick = 2,
                color = "red")
for pos, y, err in zip(x - width/2, group_21, group_21_v):
    ax2.errorbar(pos, y, err, lw = 1,
                capsize = 4, capthick = 2,
                color = "black")

for pos, y, err in zip(x + width/2, group_15, group_15_v):
    ax2.errorbar(pos, y, err, lw = 1,
                capsize = 4, capthick = 2,
                color = "red")

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(30, 50)  # outliers only
ax2.set_ylim(0, 10)  # most of the data
gridspec_kw={'height_ratios': [1,2]}

# hide the spines between ax1 and ax2
ax1.spines.bottom.set_visible(False)
ax2.spines.top.set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  
ax2.xaxis.tick_bottom()

# Only show ticks on the left and bottom spines
ax1.yaxis.set_ticks_position('left')
 
# turn towards the cut-out slanted lines.
# We create line objects in axes coordinates, in which (0,0), (0,1),
# (1,0), and (1,1) are the four corners of the axes.
# The slanted lines themselves are markers at those locations, such that the
# lines keep their angle and position, independent of the axes size or scale
# Finally, we need to disable clipping.

d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)
    
ax2.set_ylabel('%PD-1$^+$ of total CDB$^+$')
ax2.set_xticks(x, group_labels, rotation = 45)
ax1.legend(loc='upper right',frameon=False)

# add footnode of p values of different groups performen by un-paired t test
plt.text(-1,-4,"Significance of first group is %.2f." % group_significance[0])
plt.text(-1,-4.7,"Significance of second group is %.2f." % group_significance[1])
plt.text(-1,-5.3,"Significance of third group is %.2f." % group_significance[2])
plt.text(-1,-6.0,"Significance of fourth group is %.2f." % group_significance[3])
plt.text(-1,-6.7,"Significance of fifth group is %.2f." % group_significance[4])
plt.text(-1,-7.3,"Significance of last group is %.2f." % group_significance[5])

# set a foot node 
ax1.set_title('Figure5C.quantification on activated CD8+ OT-I T cells \n tests used: t test')

plt.savefig("Figure_5C")
```
